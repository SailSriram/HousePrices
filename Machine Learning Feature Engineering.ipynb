```python
#Address Missing Values for categorical features
features_nan = [feature for feature in dataset.columns if dataset[feature].isnull().sum() > 1 and dataset[feature].dtypes == 'O']
```

```python
for feature in features_nan:
    print("{}: {}% missing values".format(feature, np.round(dataset[feature].isnull().mean(), 4)))
```

```python
#Replace missing values with data label
def replace_cat_feature(dataset, features_nan):
    data = dataset.copy()
    data[features_nan] = data[features_nan].fillna('Missing Value')
    return data
dataset = replace_cat_feature(dataset, features_nan)
dataset[features_nan].isnull().sum()
```

```python
dataset.head()
```

```python
numerical_with_nan = [feature for feature in dataset.columns if dataset[feature].isnull().sum() > 1 and dataset[feature].dtypes != 'O']
for feature in numerical_with_nan:
    print("{}: {}% missing values".format(feature, np.round(dataset[feature].isnull().mean(), 4)))
```

```python
for feature in numerical_with_nan:
    median_value = dataset[feature].median()
    #Create a new feature to capture nan values
    dataset[feature + 'nan'] = np.where(dataset[feature].isnull(), 1, 0)
    dataset[feature] = dataset[feature].fillna(median_value)
dataset[numerical_with_nan].isnull().sum()
    
```

```python
dataset.head(50)
```

```python
for feature in ['YearBuilt', 'YearRemodAdd', 'GarageYrBlt']:
    dataset[feature] = dataset['YrSold'] - dataset[feature]
```

```python
#Feature Engineering for Test Data: Address missing values in categorical data
testdf = pd.read_csv("test.csv")
```

```python
categorical_nan_features = [feature for feature in testdf.columns if testdf[feature].isnull().sum() > 1 and testdf[feature].dtypes == 'O']
for feature in categorical_nan_features:
    print("{} has {}% nan values".format(feature, np.round(testdf[feature].isnull().mean(), 4)))
```

```python
def replace_null_values(testdf, categories):
    data = testdf.copy()
    data[categories] = data[categories].fillna('Missing Value')
    return data

testdf = replace_null_values(testdf, categorical_nan_features)
testdf[categorical_nan_features].isnull().sum()
```

```python
#Replace missing values in numerical features
numerical_nan_features = [feature for feature in testdf.columns if testdf[feature].isnull().sum() > 1 and testdf[feature].dtypes != 'O']
for feature in numerical_nan_features:
    print("{} has {}% nan values".format(feature, np.round(testdf[feature].isnull().mean(), 4)))
```

```python
for feature in numerical_nan_features:
    median_value = testdf[feature].median()
    #Creating new columns with 1s as null and 0s and non null for more info
    testdf[feature + 'nan'] = np.where(testdf[feature].isnull(), 1, 0)
    testdf[feature] = testdf[feature].fillna(median_value)
testdf[numerical_nan_features].isnull().sum()
    
```

```python
#Adressing missing values in date time feature
for feature in ['YearBuilt', 'YearRemodAdd', 'GarageYrBlt']:
    testdf[feature] = testdf['YrSold'] - testdf[feature]
```
